```python
import pandas as pd

df = pd.read_csv("/content/Datadata/Graphic - multiTimeline (9).csv.csv")
values = df['Graphic'].tolist()
print(values)

# take 12 values
# take average of those values
# store

import numpy as np

data = np.array(values)

# number of full chunks
n = len(data) // 12

# take only complete chunks
trimmed = data[:n*12]

averages = trimmed.reshape(n, 12).mean(axis=1)

for avg in averages:
    print(round((avg), 2))

```
